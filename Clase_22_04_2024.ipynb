```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv("casasboston.csv")

print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO CHARLES",
    "RM": "N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})

print(df)
```

```python
df.N_HABITACIONES_MEDIO.plot.hist(
    title="Frecuencia del promedio de número de habitaciones",
    xlabel="Promedio de número de habitaciones ",
    ylabel="Frecuencia")
plt.show()
```

```python
df.plot.scatter(
    x="VALOR_MEDIANO",
    y="INDICE_CRIMEN",
    alpha=0.5, c='red',
    title="Gráfico de dispersión",
    ylabel="Índice de crimen",
    xlabel="Promedio")
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()

valor_por_ciudad.head(10).plot.barh(
    title="Media de asesinatos por ciudad",
    xlabel="Promedio de asesinatos",
    ylabel="Ciudades",
    color="skyblue")
plt.show()
```
